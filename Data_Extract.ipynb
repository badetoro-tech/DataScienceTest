## Data Cleaning

```python
def missing_values_table(df):
    # Get the count of missing values per column
    mis_val = df.isnull().sum()
    
    # Calculate the percentage of missing values per column
    mis_val_percent = 100 * df.isnull().sum() / len(df)
    
    # Combine the two previous dataframes into a single one
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
    
    # Rename the columns of the new dataframe
    mis_val_table_ren_columns = mis_val_table.rename(
        columns={0: 'Missing Values', 1: '% of Total Values'})
    
    # Keep only the columns that have missing values and sort them in descending order by percentage
    mis_val_table_ren_columns = mis_val_table_ren_columns[
        mis_val_table_ren_columns.iloc[:, 1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
    
    # Print the number of columns in the original dataframe and the number of columns that have missing values
    print("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
        "There are " + str(mis_val_table_ren_columns.shape[0]) +
        " columns that have missing values.")
    
    # Return the dataframe with missing values
    return mis_val_table_ren_columns
```

```python
# Assuming your DataFrame is named 'merged_df' and the column is named 'Your_Column_Name'
column_name = 'Metrics'

# Mapping function to rename the values based on the prefix
def rename_values(value):
    prefix = value.split('_£m')[0]  # Extract the prefix (e.g., 'NWP', 'SCR', 'EoF_for_SCR')
    return prefix.rstrip('0123456789')  # Remove trailing digits

# Apply the mapping function to the specified column
df[column_name] = df[column_name].apply(rename_values)
```

```python
# Calculate the total GWP for each firm across all years
total_gwp = df[df['Metrics'] == 'GWP'].set_index('Year').filter(like='Firm').sum(axis=0)

# Identify the top ten firms based on GWP
top_ten_firms_gwp = total_gwp.nlargest(10)
print("Top ten firms based on GWP:")
print(top_ten_firms_gwp)

# Plotting the top 10 firms based on GWP
plt.figure(figsize=(10, 6))
top_ten_firms_gwp.plot(kind='bar', color='blue')
plt.title('Top 10 Firms Based on Gross Written Premium (GWP)')
plt.xlabel('Firms')
plt.ylabel('Total GWP')
plt.show()
```

```python
# Calculate the total NWP for each firm across all years
total_nwp = df[df['Metrics'] == 'NWP'].set_index('Year').filter(like='Firm').sum(axis=0)

# Identify the top ten firms based on NWP
top_ten_firms_nwp = total_nwp.nlargest(10)
print("Top ten firms based on NWP:")
print(top_ten_firms_nwp)

# Plotting the top 10 firms based on NWP
plt.figure(figsize=(10, 6))
top_ten_firms_nwp.plot(kind='bar', color='cyan')
plt.title('Top 10 Firms Based on Net Written Premium (NWP)')
plt.xlabel('Firms')
plt.ylabel('Total NWP')
plt.show()
```

```python
# Calculate the total SCR_coverage_ratio for each firm across all years
total_scr = df[df['Metrics'] == 'SCR_coverage_ratio'].set_index('Year').filter(like='Firm').sum(axis=0)

# Identify the top ten firms based on SCR Coverage Ratio
top_ten_firms_scr = total_scr.nlargest(10)
print("Top ten firms based on SCR Coverage Ratio:")
print(top_ten_firms_scr)

# Plotting the top 10 firms based on SCR Coverage Ratio
plt.figure(figsize=(10, 6))
top_ten_firms_scr.plot(kind='bar', color='blue')
plt.title('Top 10 Firms Based on SCR Coverage Ratio')
plt.xlabel('Firms')
plt.ylabel('Total SCR Coverage Ratio')
plt.show()
```

```python
df['Year'] = df['Year'].astype(int)

# Calculate the total Gross claims incurred for each firm across all years
total_gci = df[df['Metrics'] == 'Gross_claims_incurred'].set_index('Year').filter(like='Firm').sum(axis=0)

# Identify the top ten firms based on Gross claims incurred
top_ten_firms_gci = total_gci.nlargest(10)
print("Top ten firms based on Gross claims incurred:")
print(top_ten_firms_gci)

# Plotting the top 10 firms based on Gross claims incurred (Line Chart)
plt.figure(figsize=(12, 6))

# Plotting the top 10 firms based on SCR Coverage Ratio
plt.figure(figsize=(10, 6))
top_ten_firms_gci.plot(kind='bar', color='gray')
plt.title('Top 10 Firms Based on Gross claims')
plt.xlabel('Firms')
plt.ylabel('Total Gross claims')
plt.show()
```

```python
for firm in top_ten_firms_gci.index:
    # Update the column name if needed
    plt.plot(df[df['Metrics'] == 'Gross_claims_incurred'].set_index('Year')[firm], label=firm)

plt.title('Top 10 Firms Based on Gross claims incurred Over Years')
plt.xlabel('Year')
plt.ylabel('Gross claims incurred')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Calculate the total Net combined ratio for each firm across all years
total_ncr = df[df['Metrics'] == 'Net_combined_ratio'].set_index('Year').filter(like='Firm').sum(axis=0)

# Identify the top ten firms based on Net combined ratio
top_ten_firms_ncr = total_ncr.nsmallest(10)
print("Top ten firms based on Net combined ratio:")
print(top_ten_firms_ncr)

# Plotting the top 10 firms based on Net combined ratio
plt.figure(figsize=(12, 6))

# Plotting the top 10 firms based on Net combined ratio
plt.figure(figsize=(10, 6))
top_ten_firms_ncr.plot(kind='bar', color='gray')
plt.title('Top 10 Firms Based on Net combined ratio')
plt.xlabel('Firms')
plt.ylabel('Net combined ratio')
plt.show()
```

```python
# Calculate the total assets for each firm across all years
total_assets = df[df['Metrics'] == 'Total_assets'].set_index('Year').filter(like='Firm').sum(axis=0)

# Identify the top ten firms based on total assets
top_ten_firms_assets = total_assets.nlargest(10)
print("Top ten firms based on total assets:")
print(top_ten_firms_assets)

# Convert values to £ million
top_ten_firms_assets_million = top_ten_firms_assets / 1e6

# Plotting the top 10 firms based on total assets in £ million
plt.figure(figsize=(10, 6))
top_ten_firms_assets_million.plot(kind='bar', color='green')
plt.title('Top 10 Firms Based on Total Assets')
plt.xlabel('Firms')
plt.ylabel('Total Assets (£ million)')
plt.tick_params(axis='y', labelrotation=0)
plt.show()
```
